```python
# Import des modules nécessaires
from pymongo import MongoClient
import polars as pl

# Config affichage polars
pl.Config.set_tbl_rows(50)     # nombre de lignes affichées
pl.Config.set_tbl_cols(20)     # nombre de colonnes affichées
```

```python
# Connexion à la base MongoDB

client = MongoClient("mongodb://localhost:27017")

db = client["noscites"]
collection = db["logements"]
```

```python
# Vérification du bon import de la collection "logements"
print(collection.count_documents({}))
```

```python
# Transformation de la collection importée en liste normalisée
docs = list(collection.find())
for d in docs:
    d["_id"] = str(d["_id"])

LIST_FIELDS = ["amenities","host_verifications"]  # les champs listes à nettoyer

# remplace les listes vides par None
for d in docs:
    for f in LIST_FIELDS:
        if f in d and d[f] == []:
            d[f] = None
```

```python
# Transforme la liste en Dataframe
df = pl.DataFrame(docs, infer_schema_length=10_000)
```

```python
# Vérification du DataFrame
print(df.shape)
print(df.columns)
df.head()
```

```python
(
    df
    .select("id","number_of_reviews")
    .sort("number_of_reviews", descending=True)
    .limit(5)
)
```

```python
(
    df
    .select(pl.col("instant_bookable").cast(pl.Int8).mean().alias("proportion_instant_bookable"))
    
)
```

```python
(
    df
    .group_by("room_type")
    .agg(
        pl.col("instant_bookable")
        .cast(pl.Int8)
        .mean()
        .alias("proportion_instant_bookable")
    )
)
```

```python
# Calculer le taux de réservation moyen par mois par type de logement
(
    df
    .with_columns(
        (
            1 - (pl.col("availability_30") / 30)
        ).alias("booking_rate_30d")
    )
    .group_by("room_type")
    .agg(
        pl.col("booking_rate_30d")
        .mean()
        .alias("avg_monthly_booking_rate")
    )
    .sort("avg_monthly_booking_rate", descending=True)
)
```

```python
# Calculer la médiane des nombre d’avis pour tous les logements
(
    df
    .select(pl.col("number_of_reviews").median().alias("Total number of reviews - Median"))
)
```

```python
# Calculer la médiane des nombre d’avis par catégorie d’hôte

(
    df
    .group_by("host_is_superhost")
    .agg(pl.col("number_of_reviews").median().alias("Number of reviews - Median"))
)
```

```python
# Calculer la densité de logements par quartier de Paris

(
    df
    .group_by("neighbourhood_cleansed")
    .agg(pl.len().alias("n_accomodations"))
    .with_columns((pl.col("n_accomodations")/pl.col("n_accomodations").sum()).alias("accomodation_density"))
    .sort("accomodation_density", descending=True)
    .rename({"neighbourhood_cleansed": "paris_neighbourhood"})
)
```

```python
# Identifier les quartiers avec le plus fort taux de réservation par mois

(
    df
    .with_columns(
        (
            1 - (pl.col("availability_30") / 30)
        ).alias("booking_rate_30d")
    )
    .group_by("neighbourhood_cleansed")
    .agg(
        pl.col("booking_rate_30d")
        .mean()
        .alias("avg_monthly_booking_rate")
    )
    .sort("avg_monthly_booking_rate", descending=True)
    .rename({"neighbourhood_cleansed": "paris_neighbourhood"})
)
```
